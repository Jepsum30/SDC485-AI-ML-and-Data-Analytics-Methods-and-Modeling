```python
from IPython.display import Markdown, display
import numpy as np
import matplotlib.pyplot as plt

display(Markdown("""
## STEP 2 – Imports  
**Date:** 05/05/2026  
**Student ID:** JOSZIM1500
"""))
```

```python
display(Markdown("""
## STEP 3 – Loan Payment Function  
**Date:** 05/05/2026  
**Student ID:** JOSZIM1500
"""))

def monthly_payment(P, annual_rate, years):
    r = annual_rate / 12
    n = years * 12
    if r == 0:
        return P / n
    M = P * r * (1 + r)**n / ((1 + r)**n - 1)
    return M
```

```python
display(Markdown("""
## STEP 4 – Scenario Calculations  
**Date:** 05/05/2026  
**Student ID:** JOSZIM1500
"""))

P = 300_000

scenarios = {
    "Scenario 1\n3% - 15 yrs": {"rate": 0.03, "years": 15},
    "Scenario 2\n5% - 30 yrs": {"rate": 0.05, "years": 30},
    "Scenario 3\n8% - 40 yrs": {"rate": 0.08, "years": 40},
}

results = []

for name, params in scenarios.items():
    rate = params["rate"]
    years = params["years"]
    M = monthly_payment(P, rate, years)
    n = years * 12
    total = M * n
    results.append({"name": name, "monthly": M, "total": total})

for r in results:
    print(r["name"])
    print(f"  Monthly Payment: ${r['monthly']:.2f}")
    print(f"  Total Repayment: ${r['total']:.2f}")
    print("-" * 40)
```

```python
display(Markdown("""
## STEP 5 – Bar Chart Visualization  
**Date:** 05/05/2026  
**Student ID:** JOSZIM1500
"""))

names = [r["name"] for r in results]
totals = [r["total"] for r in results]

plt.figure(figsize=(8,5))
bars = plt.bar(names, totals, color=["#4c72b0", "#55a868", "#c44e52"])
plt.title("Total Loan Repayment by Scenario")
plt.ylabel("Total Repayment ($)")
plt.xticks(rotation=15)

for bar, total in zip(bars, totals):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f"${total:,.0f}", ha="center", va="bottom")

plt.tight_layout()
plt.show()
```

```python
display(Markdown("""
## STEP 6 – Sensitivity Analysis  
**Date:** 05/05/2026  
**Student ID:** JOSZIM1500
"""))

years_sa = 30
rates_sa = np.arange(0.02, 0.091, 0.01)

totals_sa = []

for rate in rates_sa:
    M = monthly_payment(P, rate, years_sa)
    total = M * years_sa * 12
    totals_sa.append(total)

for rate, total in zip(rates_sa, totals_sa):
    print(f"Rate: {rate*100:.1f}% | Total: ${total:,.2f}")
```

```python
display(Markdown("""
## STEP 7 – Sensitivity Analysis Visualization  
**Date:** 05/05/2026  
**Student ID:** JOSZIM1500
"""))

plt.figure(figsize=(8,5))
plt.plot(rates_sa * 100, totals_sa, marker="o", color="#4c72b0")
plt.title("Sensitivity of Total Repayment to Interest Rate (30-Year Term)")
plt.xlabel("Interest Rate (%)")
plt.ylabel("Total Repayment ($)")
plt.grid(True, linestyle="--", alpha=0.5)

for x, y in zip(rates_sa * 100, totals_sa):
    plt.text(x, y, f"${y/1000:,.0f}k", ha="center", va="bottom", fontsize=8)

plt.tight_layout()
plt.show()
```
