```python
# Import necessary libraries
import pandas as pd
import numpy as np
from datetime import datetime
from IPython.display import display, Markdown
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Student Information
todays_date = datetime.now().strftime("%Y-%m-%d")
student_id = "JOSZIM1500"

display(Markdown("# Data Preprocessing on Breast Cancer Dataset"))
display(Markdown(f"**Date:** {todays_date}  \n**Student ID:** {student_id}"))

```

```python
# ===================================================================
# Step 1: Load and Explore the Dataset
# ===================================================================
display(Markdown("### Step 1: Load and Explore the Dataset"))
display(Markdown(f"**Date:** {todays_date}  \n**Student ID:** {student_id}"))

# Load the dataset
df = pd.read_csv('breastCancerDataset.csv')

display(Markdown("**1. Dataset Information:**"))
df.info()

display(Markdown("**2. Statistical Summary:**"))
display(df.describe())

display(Markdown("**3. First 5 Rows of the Dataset:**"))
display(df.head())
```

```python
# ===================================================================
# Step 2: Preprocess the Data
# ===================================================================
display(Markdown("### Step 2: Preprocess the Data"))
display(Markdown(f"**Date:** {todays_date}  \n**Student ID:** {student_id}"))

# 1. Check missing values
display(Markdown("**Handling Missing Values:**"))
display(Markdown(f"Total missing values: {df.isnull().sum().sum()}"))
df = df.dropna()

# 2. Remove duplicate rows
display(Markdown("**Removing Duplicate Rows:**"))
display(Markdown(f"Duplicates before removal: {df.duplicated().sum()}"))
df = df.drop_duplicates()
display(Markdown(f"Duplicates after removal: {df.duplicated().sum()}"))

# 3. Encode Diagnosis (M = 1, B = 0)
df['Diagnosis'] = df['Diagnosis'].map({'M': 1, 'B': 0})

# 4. Standardize concavity1 and concavity3
scaler = StandardScaler()
df[['concavity1', 'concavity3']] = scaler.fit_transform(df[['concavity1', 'concavity3']])

display(Markdown("**Standardized Features (First 5 rows):**"))
display(df[['concavity1', 'concavity3']].head())

# Prepare X and y
X = df.drop(['ID', 'Diagnosis'], axis=1)
y = df['Diagnosis']

# Train-Test Split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

display(Markdown("**Train-Test Split Completed:**"))
display(Markdown(f"Training samples: {X_train.shape[0]}"))
display(Markdown(f"Testing samples: {X_test.shape[0]}"))
```

```python
# ===================================================================
# Step 3: Implement SVM
# ===================================================================
display(Markdown("### Step 3: Implement SVM"))
display(Markdown(f"**Date:** {todays_date}  \n**Student ID:** {student_id}"))

svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

display(Markdown("**SVM Model Results:**"))
display(Markdown(f"**Accuracy Score:** {accuracy_score(y_test, y_pred_svm):.4f}"))
display(Markdown("**Classification Report:**"))
print(classification_report(y_test, y_pred_svm))
```

```python
# ===================================================================
# Step 4: GridSearchCV for SVM Hyperparameter Tuning
# ===================================================================
display(Markdown("### Step 4: GridSearchCV for SVM Hyperparameter Tuning"))
display(Markdown(f"**Date:** {todays_date}  \n**Student ID:** {student_id}"))

param_grid_svm = { 'C': [0.1, 1, 10, 100], 'kernel': ['linear', 'rbf', 'poly'] }

grid_svm = GridSearchCV(SVC(random_state=42), param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(X_train, y_train)

display(Markdown(f"**Best Parameters:** {grid_svm.best_params_}"))
display(Markdown(f"**Best CV Score:** {grid_svm.best_score_:.4f}"))

best_svm = grid_svm.best_estimator_
y_pred_best_svm = best_svm.predict(X_test)

display(Markdown("**Tuned SVM Results:**"))
display(Markdown(f"**Accuracy Score:** {accuracy_score(y_test, y_pred_best_svm):.4f}"))
display(Markdown("**Classification Report:**"))
print(classification_report(y_test, y_pred_best_svm))
```

```python
# ===================================================================
# Step 5: Implement XGBoost
# ===================================================================
display(Markdown("### Step 5: Implement XGBoost"))
display(Markdown(f"**Date:** {todays_date}  \n**Student ID:** {student_id}"))

xgb_model = XGBClassifier(random_state=42, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)

display(Markdown("**XGBoost Model Results:**"))
display(Markdown(f"**Accuracy Score:** {accuracy_score(y_test, y_pred_xgb):.4f}"))
display(Markdown("**Classification Report:**"))
print(classification_report(y_test, y_pred_xgb))
```
